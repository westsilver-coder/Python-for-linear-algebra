```python
# import modules
import numpy as np
```

```python
v1 = np.array([4,5,6])
v2 = np.array([-4,0,5])

np.dot(v1,v2)
```

```python
v3 = np.array([-4,3,1])

np.dot(v2,v3)
```

```python
v4 = np.array([0,0,0])
np.dot(v3,v4)
```

```python
v3 = np.array([-4,3,1])
v5 = np.array([2,3,-1])
np.dot(v3,v5)
# vector v3 is orthogonal to vector v5
```

```python
# assignment
# 1) come up with 3 2D vectors, two of them are orthogonal, but neither is orthogonal to the third
# 2) plot all three vectors -- make sure the axes are square and equal.

import matplotlib.pyplot as plt

vec1 = np.array([1,2,3])
vec2 = np.array([-3,3,-1])
vec3 = np.array([2,2,2])

plt.plot([0,vec1[0]],[0,vec1[1]],'rs-',label='vec1')
plt.plot([0,vec2[0]],[0,vec2[1]],'gs-',label='vec2')
plt.plot([0,vec3[0]],[0,vec3[1]],'bs-',label='vec3')

plt.axis('square')
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.grid()
plt.legend()
plt.show()
```
