```python
# import modules
import matplotlib.pyplot as plt
import numpy as np
```

```python
vec = np.array([3, 4, 5, 2])
s = 3

vec*s
```

```python
vec2d=np.array([1,2])
s1 = 2
s2 = .5
s3 = -1

# indexing
vec2d[0]
```

```python
plt.plot([0,vec2d[0]],[0,vec2d[1]],'bs-')
```

```python
plt.plot([0,vec2d[0]],[0,vec2d[1]], 'bs-', label='v')
plt.plot([0,s1*vec2d[0]],[0,s1*vec2d[1]],'ro-',label='v*s1')
plt.plot([0,s2*vec2d[0]],[0,s2*vec2d[1]],'kp-',label='v*s2')
plt.plot([0,s3*vec2d[0]],[0,s3*vec2d[1]],'g*-',label='v*s3')

plt.axis('square')
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.grid()
plt.legend()
plt.show()
```

```python
# assignment!
# create two vectors (2D vectors)
# plot each vector
# plot v1+v2, v1-v2, v1*4 + v2/2

v1=np.random.randint(0,4,size=2)
v2=np.random.randint(0,4,size=2)
print(v1)
print(v2)

plt.plot([0,v1[0]],[0,v1[1]],'ro-', label='v1')
plt.plot([0,v2[0]],[0,v2[1]],'bo-', label='v2')

plt.plot([0,v1[0]+v2[0]],[0,v1[1]+v2[1]],'ko-', label='v1+v2')
plt.plot([0,v1[0]-v2[0]],[0,v1[1]-v2[1]],'yo-', label='v1-v2')
plt.plot([0,4*v1[0]-v2[0]/2],[0,4*v1[1]-v2[1]/2],'go-', label='v1*4-v2/2')
plt.grid()
plt.legend()
plt.show()
```
